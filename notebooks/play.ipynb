```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq


filename = "ratings"
df = pd.read_csv(f"../ml-latest/{filename}.csv")
df = df.sort_values("timestamp", ascending=False).reset_index(drop=True)
display(df)
df.info()
```

```python
table = pa.Table.from_pandas(df)

pq.write_to_dataset(
    table,
    root_path=f"../data/movielens/{filename}.parquet",
    partition_cols=['rating'],
    compression='gzip' 
)
```
